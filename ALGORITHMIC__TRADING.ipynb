# Amazon cumulative returns and drawdown

```python

amzn = yf.download("AMZN").Close.to_frame()
amzn.head()
```

```python
amzn["d_returns"] = np.log(amzn.div(amzn.shift(1)))
amzn.dropna(inplace=True)
amzn["cummreturns"] = amzn["d_returns"].cumsum().apply(np.exp)
amzn["cummmax"] = amzn["cummreturns"].cummax()
amzn["drawdown"] = amzn["cummmax"] - amzn["cummreturns"]
amzn["drawdown%"] = amzn["drawdown"] / amzn["cummmax"]
amzn[["cummreturns", "cummmax"]].plot(figsize=(12, 8), title="AMZN Buy and Hold + Cummmax", fontsize=12)
plt.show()
```

# SMA strategy function

```python

def test_strategy(stock, start, end, SMA):
    df = yf.download(stock, start=start, end=end)
    data = df["Close"].to_frame()
    data["returns"] = np.log(data["Close"].div(data["Close"].shift(1)))
    data["SMA_S"] = data["Close"].rolling(int(SMA[0])).mean()
    data["SMA_L"] = data["Close"].rolling(int(SMA[1])).mean()
    data.dropna(inplace=True)
    data["position"] = np.where(data["SMA_S"] > data["SMA_L"], 1, -1)
    data["strategy"] = data["returns"] * data["position"].shift(1)
    data.dropna(inplace=True)
    ret = np.exp(data["strategy"].sum())
    std = data["strategy"].std() * np.sqrt(252)
    return ret, std
```

```python

print(test_strategy("GOOGL", "2000-01-01", "2020-01-01", (50, 200)))
```

# SMA strategy class

```python

class SMABacktester:
    def __init__(self, symbol, SMA_S, SMA_L, start, end):
        self.symbol = symbol
        self.SMA_S = SMA_S
        self.SMA_L = SMA_L
        self.start = start
        self.end = end
        self.results = None
        self.get_data()

    def get_data(self):
        df = yf.download(self.symbol, start=self.start, end=self.end)
        data = df.Close.to_frame()
        data["returns"] = np.log(data.Close.div(data.Close.shift(1)))
        data["SMA_S"] = data.Close.rolling(self.SMA_S).mean()
        data["SMA_L"] = data.Close.rolling(self.SMA_L).mean()
        data.dropna(inplace=True)
        self.data2 = data
        return data

    def test_results(self):
        data = self.data2.copy().dropna()
        data["position"] = np.where(data["SMA_S"] > data["SMA_L"], 1, -1)
        data["strategy"] = data["returns"] * data.position.shift(1)
        data.dropna(inplace=True)
        data["returns_bh"] = data["returns"].cumsum().apply(np.exp)
        data["returns_strategy"] = data["strategy"].cumsum().apply(np.exp)
        perf = data["returns_strategy"].iloc[-1]
        outperf = perf - data["returns_bh"].iloc[-1]
        self.results = data
        ret = np.exp(data["strategy"].sum())
        std = data["strategy"].std() * np.sqrt(252)
        return round(perf, 6), round(outperf, 6)

    def plot_results(self):
        if self.results is None:
            print("Run the test please")
        else:
            title = "{} | SMA Short={} | SMA Long={}".format(self.symbol, self.SMA_S, self.SMA_L)
            self.results[["returns_bh", "returns_strategy"]].plot(title=title, figsize=(12, 8), color=["blue", "orange"])
            plt.title("SMA Strategy Backtest Results", fontsize=15, color='darkblue')
            plt.xlabel("Date", fontsize=12)
            plt.ylabel("Cumulative Returns", fontsize=12)
            plt.grid(True, which='both', linestyle='--', linewidth=0.5)
            plt.show()
```

```python
sma_tester = SMABacktester("GOOGL", 50, 100, "2000-01-01", "2020-01-01")
```

```python
print(sma_tester.test_results())
```

```python
sma_tester.plot_results()
```
